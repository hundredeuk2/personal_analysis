```python
incheon_df = pd.read_csv(r'아파트(매매)__실거래가_20210310160909.csv', encoding="cp949")
```

```python
incheon_df['거래금액(만원)'] = incheon_df['거래금액(만원)'].str.replace(pat=",", repl='', regex=False)
incheon_df.head()
```

```python
incheon_df['거래금액(만원)'] = pd.to_numeric(incheon_df['거래금액(만원)'])
incheon_df.head()
```

```python
price_mean_df = incheon_df.groupby(["계약년월","시군구"])['거래금액(만원)'].mean().reset_index()
price_mean_df.head()
```

```python
yeonsugu_df = price_mean_df.loc[price_mean_df['시군구']. str.contains("연수구")]
yeonsugu_df.head()
```

```python
yeonsugu_mean_df = yeonsugu_df.groupby(['계약년월'])['거래금액(만원)'].mean().reset_index()
yeonsugu_mean_df.head()
```

```python
price_count_df = incheon_df.groupby(["계약년월","시군구"])['거래금액(만원)'].count().reset_index()
price_count_df.head()
```

```python
yeonsugu_df = price_count_df.loc[price_mean_df['시군구']. str.contains("연수구")]
yeonsugu_df.head()
```

```python
yeonsugu_count_df = yeonsugu_df.groupby(['계약년월'])['거래금액(만원)'].sum().reset_index()
yeonsugu_count_df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15,15))
plt.bar(data=yeonsugu_count_df, x="계약년월",height = '거래금액(만원)')
sns.lineplot(data=yeonsugu_mean_df, x="계약년월", y="거래금액(만원)")

plt.show()
```

```python
gu_list = ["연수구","남동구","서구","중구","동구","미추홀구","부평구","계양구"]
```

```python
for i in gu_list:
    a = price_mean_df.loc[price_mean_df['시군구']. str.contains(i)]
    sns.lineplot(data=a, x="계약년월", y="거래금액(만원)")
    
plt.show()
```

```python
for i in gu_list:
    a = price_count_df.loc[price_mean_df['시군구']. str.contains(i)]
    b = a.groupby(['계약년월'])['거래금액(만원)'].sum().reset_index()
    plt.bar(data=b, x="계약년월",height = '거래금액(만원)')

plt.legend()
plt.show()
```
